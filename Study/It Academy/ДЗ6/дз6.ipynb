```python
# Задача 1
A = 2 / 6  # Вероятность события A
B= 1 - A  # Вероятность события B

print(f"Вероятность A (1 или 2): {A}")
print(f"Вероятность B (не 1 или 2): {B}")

# Распределение дискретной случайной величины X
distribution_of_DRV =[ [1, 2, 3, 4, 5, 6],[1/6] * 6] # Равномерное распределение
print(f"Вероятности для возможных исходов X:\n{distribution_of_DRV[0]}\n{distribution_of_DRV[1]}")
```

```python
# Задача 2
P_single = 3 / 6  # Вероятность выпадения нечетного числа на одном кубике
P_both = P_single * P_single  # Вероятность, что оба будут нечетными

print(f"Вероятность, что оба кубика дадут нечетное число: {P_both}")
```

```python
# Задача 3
P_A = 3 / 6  # Вероятность выпадения нечетного числа
P_B = 1 / 6  # Вероятность 5-ки на втором кубике

P_A_and_B = P_A * P_B  # Вероятность одновременного наступления событий A и B

print(f"Вероятность, что произошло и A и B: {P_A_and_B}")
```

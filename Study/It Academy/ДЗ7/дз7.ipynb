```python
from tensorflow import keras as k
```

```python
model_01 = k.Sequential()
model_01.add(k.layers.Conv2D(32,(5, 5), activation = 'relu', input_shape = (50, 50, 1)))
model_01.add(k.layers.MaxPool2D())
model_01.add(k.layers.Conv2D(64,(4, 4), activation = 'relu'))
model_01.add(k.layers.MaxPool2D())
model_01.add(k.layers.Flatten())
model_01.add(k.layers.Dense(500, activation = 'sigmoid'))
model_01.add(k.layers.Dense(10, activation = 'softmax'))

model_01.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])
print(model_01.summary())
```

Я решил использовать два сверточных слоя, так как изображения и так достаточно маленькие. Размеры ядер я сначала указал 5 на 5, чтобы выявить локальные признаки. На втором слое сделал 4 на 4 для обобщения локальных признаков. Количество ядер увеличивал с 32 до 64, чтобы подчеркнуть присутствие различных признаков в изображении.

```python
model_02 = k.Sequential()
model_02.add(k.layers.Conv2D(32,(5, 5), activation = 'relu', input_shape = (150, 150, 3)))
model_02.add(k.layers.MaxPool2D())
model_02.add(k.layers.Conv2D(64,(4, 4), activation = 'relu'))
model_02.add(k.layers.MaxPool2D())
model_02.add(k.layers.Conv2D(128,(3, 3), activation = 'relu'))
model_02.add(k.layers.MaxPool2D())
model_02.add(k.layers.Flatten())
model_02.add(k.layers.Dense(1000, activation = 'sigmoid'))
model_02.add(k.layers.Dense(1, activation = 'sigmoid'))

model_02.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
print(model_02.summary())
```

В этой задаче я решил использовать 3 сверточных слоя, так как изображения достаточно большие. Количество ядер я постепенно увеличивал с 32 до 64, а их размер уменьшал для обобщения признаков.Сначала использовал ядра размером 5 на 5, затем 4 на 4 и 3 на 3. Придерживался аналогичной стратегии, что и в первом задании.

```python
model_03 = k.Sequential([
    k.layers.Input(shape=(60,)),
    k.layers.Dense(128, activation='relu'),
    k.layers.Dense(64, activation='relu'),
    k.layers.Dense(1)
])

model_03.compile(optimizer='adam', loss='mean_squared_error')
```

Я взял два скрытых слоя для обобщения данных. Сначала взял 128 нейронов для выявления зависимостей в данных, а во втором слое уменьшил до 64 для обобщения.
